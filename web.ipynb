```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://www.amazon.com/s?k=monitor&sprefix=moni%2Caps%2C345&ref=nb_sb_ss_ts-doa-p_2_4"
```

```python
# Headers for request
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
```

```python
# HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
type(webpage.content)
```

```python
# Soup Object containing all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal")
```

```python
link = links[0].get('href')
```

```python
product_link = "https://www.amazon.com" + link
```

```python
product_link
```

```python
new_webpage = requests.get(product_link, headers=HEADERS)
```

```python
new_webpage
```

```python
# Soup Object containing all data
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
 # Fetch links as List of Tag Objects
new_soup.find("span", attrs={"id": "productTitle"}).text.strip()
```

```python
new_soup.find("span", attrs={"class": "a-price a-text-price"}).find("span", attrs={"class": "a-offscreen"}).text.strip()
```
